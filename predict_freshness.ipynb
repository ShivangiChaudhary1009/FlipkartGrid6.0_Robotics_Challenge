```python
from PIL import Image

def preprocess_image(image_path):
    img = Image.open(image_path).convert('RGB')

    # Define the transformations (same as those used in training)
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean = 0, std = 1)
    ])

    # Apply transformations
    img = transform(img)

    # Add batch dimension (1, C, H, W)
    img = img.unsqueeze(0)

    return img


def predict_image(image_path):
    img = preprocess_image(image_path).to(device)

    # Disable gradient calculation (we are doing inference)
    with torch.no_grad():
        # Pass the image through the model
        output_fruit, output_fresh = model(img)

        # Get the predicted class for both outputs
        pred_fruit = torch.argmax(output_fruit, axis=1).item()
        pred_fresh = torch.argmax(output_fresh, axis=1).item()

        # Map the predictions to class names
        fruit_name = le.inverse_transform([pred_fruit])[0]  # Convert label to fruit name
        freshness = 'Fresh' if pred_fresh == 0 else 'Spoiled'

    return fruit_name, freshness
```
